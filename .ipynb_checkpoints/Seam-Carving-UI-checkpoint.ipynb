```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import matplotlib.cm as cm
import numpy as np
import copy
import cv2
from matplotlib.widgets import Slider
import matplotlib.path as mplPath
```

```python
org_img = mpimg.imread('dolphin.jpeg')
```

```python
'''
UI Code for Slider and drawing on image using Brush
'''

%matplotlib tk
brush_size = 2
img = copy.deepcopy(org_img)
rem_points = []

fig, ax = plt.subplots()
imgplot = plt.imshow(img)
plt.show()

cid = fig.canvas.mpl_connect('button_press_event', onclick)

plt.subplots_adjust(left=0.25, bottom=0.25)

axamp = plt.axes([0.1, 0.25, 0.0225, 0.63])
brush_slider = Slider(
    ax=axamp,
    label="Amplitude",
    valmin=1,
    valmax=10,
    valinit=2,
    orientation="vertical"
)
def update(val):
    global brush_size 
    brush_size = int(val) 
    fig.canvas.draw_idle()

brush_slider.on_changed(update)

def onclick(event):
    global imgplot,img,brush_size
    if event.xdata is None:
        return
    ix, iy = int(event.xdata), int(event.ydata)
    for i in range(-brush_size,brush_size+1):
        for j in range(-brush_size,brush_size+1):
            if [i+iy,j+ix] in rem_points:
                continue
            img[i+iy][j+ix] = np.array([0,255,0])
            rem_points.append([ix, iy])
    imgplot.set_data(img)
    plt.show()
```

```python
%matplotlib tk
img = copy.deepcopy(org_img)
rem_points = []

fig, ax = plt.subplots()
imgplot = plt.imshow(img)
plt.show()

def onclick(event):
    ix, iy = int(event.xdata), int(event.ydata)
    rem_points.append([ix, iy])
    
cid = fig.canvas.mpl_connect('button_press_event', onclick)
```

```python
%matplotlib inline
rem_points = np.array(rem_points)
poly_path = mplPath.Path(rem_points)
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        if(poly_path.contains_point(np.array([j,i]))):
           img[i][j] = [0,255,0]

        
imgplot = plt.imshow(img)
plt.scatter(rem_points[:,0],rem_points[:,1],c="r",s=10)
plt.show()
```
